```python
import csv
import sys
csv.field_size_limit(sys.maxsize)
```

```python
PREFIX = 'Ponyland/'
path = PREFIX + 'headhunter_sallary.txt.proper.csv'
#define MODE=PRODUCTION
```

```python
import json
with open(path + '.ids_skills' + '.json') as f:
    ids_skills = json.load(f)
```

```python
len(ids_skills)
```

```python
ids_skills = dict(ids_skills.items())
```

```python
import gensim
```

```python
def TLD(ids_skills): # a gensim TaggedDocument generator
    for id, skills in ids_skills.items():
        yield gensim.models.doc2vec.TaggedDocument(skills, id)
```

```python
#model = gensim.models.Doc2Vec(TLD(ids_skills), size=100, window=8, min_count=5, workers=8)
model = gensim.models.Doc2Vec(TLD(ids_skills), size=50, window=2, min_count=1, sample=1e-5, iter=10, workers=8)
```

```python
model.n_similarity(['1с: предприятие 8'], ['прямые продажи'])
```

```python
model.wmdistance(['1с: предприятие 8'], ['прямые продажи'])
```

```python
model.most_similar(positive=['пользователь пк'], negative=['работа в команде'])
```

```python
model.most_similar(positive=['поиск и привлечение клиентов'], negative=['телефонные переговоры'])
```

```python
def f():
    for k, v in ids_skills.items():
        yield {k: tuple(map(lambda x: tuple((float(i) for i in model[x])), v))}
```
